```python
import os
import pandas as pd
```

```python
experiments = ['square', 'random', 'chirp', 'melon']
runs = [1, 2, 3, 4]
df_real_dict = {}

for experiment in experiments:
    file_name = f'{experiment}_20251017'
    for run in runs:
        try:
            file_name_real = f'{file_name}_run{run}.csv'
            df_real = pd.read_csv(os.path.join('../data/real/processed/train/', file_name_real))
            # Store dynamically
            df_real_dict[f'{experiment}_run{run}'] = df_real
        except Exception as e:
            print(e)
            continue
```

```python
pd.DataFrame({"columns": df_real.columns}).to_excel("columns.xlsx", index=False)
```
